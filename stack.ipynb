# Stack vs Queue

Stack (LIFO)

- underlying structure: **list/dynamic array**( continouse memory)
- push = list.append(): to right/end [1,2] -> [1,2,3] **O(1)**
- pop= list.pop(): from right/end [1,2,3] -> [1,2] **O(1)**

Queue (FIFO): two ways to implement

* **deque([])** official way. why? pop is fast 
  - underlying structure: **linked list** (non continouse memory)
  - push =deque.append(): to right/end [1,2] -> [1,2,3] **O(1)**
  - pop =deque.popleft(): from left/beginning [1,2,3] ->[2,3] **O(1)**

* list: works but not recommended
  - underlying structure: dynamic array
  - push =list.append(): to right/end [1,2] -> [1,2,3] **O(1)**
  - pop =list.pop(0): from left/beginning [1,2,3] ->[2,3] **O(N)**


List/Dynamic array:
- append 0(1) 
- pop right O(1)
- pop left O(n) (Tradeoff shift all elements)
- Look up O(1) (Main reason why list uses dynamic array but not linked list)
- insert O(n) (Tradeoff shift all elements)

# Leetcode 1381-Design a Stack With Increment Operation

Description:

Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.

Implement the `CustomStack` class:

- `CustomStack(int maxSize)` Initializes the object with maxSize which is the maximum number of elements
in the stack or do nothing if the stack reached the maxSize.
- `void push(int x)` Adds x to the top of the stack if the stack hasn’t reached the maxSize.

- `int pop()`pops and returns the top of the stack or -1 if the stack is empty.

- `void inc(int k, int val)` increments the bottol of k elements of the stack by val. If there are less than k elements in the stack, increment all the elements in the stack.

# Thoughts

Use List to implement Stack: LIFO

push is append O(1)

pop is pop O(1)

inc bottom k elements are the oldest? O(K)

inc could be O(1) using extra array (pay the debt as you pop) -> did not implement
 


```python
class CustomStack():
    def __init__(self, maxSize:int)-> None:
        self._myStack=[] # initiate stack as a list
        self._size=maxSize # attribute is the max size of list
    
    def push(self,x:int) -> None:
        if len(self._myStack)< self._size: # no =
            self._myStack.append(x) # O(1) do i need to raise error if reach max

    def pop(self) -> int:
        if not self._myStack:
            return -1
        else:
            return self._myStack.pop() #O(1)

    
    def inc(self,k: int, val: int) -> None:
        # the bottom of the stacks are oldest ones, start with index 0 to k

        # if k> current length, then update k to the current length
        k=min(k,len(self._myStack))

        for i in range(k): #O(k)
            self._myStack[i]+= val
```

# Leetcode 1047- Remove All Adjacent Duplicates In String

Description:
Given a string S of lowercase letters, a duplicate removal consists of choosing two adjacent and equal letters, and removing them.

We repeatedly make duplicate removals on S until we no longer can.

Return the final string after all such duplicate removals have been made. It is guaranteed the answer is unique.

Note:
<= S.length <= 20000.
consists only of English lowercase letters.
Example:
Input:"abbaca"

Output:"ca"

Explanation:

Thoughts:

use stack time O(n) space O(n)

algorithem:

initiate a pointer i to track index at s

initiate a stack/list structure to track elements


while i<=len(s)-1:
> check if stack is empty, if so, directly add s[i] to stack 

> if not emty: 

> > check next char at s s[i] == stack.top()([-1])

> > if yes: 
 - remove stack top element 

> > else:
 - add current char to stack 

> pointer move to next

return stack

```python
def pop_duplicates(s:str)->str:
    i = 0
    myStack=[]

    while i<=len(s)-1:
        if not myStack: # if empty, add to the stack and move the pointer
            myStack.append(s[i])
        else: # if not empty, compare
            if s[i] == myStack[-1]: # if next char in s matches stack's top,pop
                myStack.pop()
            else: # else add current char to stack
                myStack.append(s[i])
        i+= 1

    return str(''.join(myStack))
```

```python
print(pop_duplicates('abbaca'))
```

# Leetcode 20-Valid Parentheses

Description:
Given a string containing just the characters "(", ")", "{", "}", "[" and "]", determine if the input string is valid.

An input string is valid if:

Open brackets must be closed by the same type of brackets.
Open brackets must be closed in the correct order.
Note:
That an empty string is also considered valid.

Example1:
Input: "( )"

Output: true

Example2:
Input: "( ) [ ] { }"

Output: true

Example3:
Input: "( ]"

Output: false

Example4:
Input: "( [ ) ]"

Output: false

Example5:
Input: "{ [ ] }"

Output: true

Thoughts:

Brute force O(N2)

Solutinon1 stack

There are two type of brackets- open and closed. The most recent open bracket(e.g. '(') must match the first added close bracket(e.g. ')'). Think in example 2 and 5. Think more in flow ranther in index.

Algorithm:

1. initiate a stack=[]

2. initiate a pointer i for s

3. a dictionary to store open/close pairs: key=open, value=closed

4. while i <=len(s)-1:

   - if stack is emty: add s[i] to stack, move pointer in s

   - if not empty:

       - if s[i] is one of the open bracket (in dictionary.keys()):

         - add to the stack, move pointer in s

       - else then it is a closed bracket, does it match stack top using dictionary pair:

            - if so: pop stack top, keeping moving the pointer i+=1

            - if not, return false

return True if we can get to this step? (but for one element: have to be checked in the beginning) 

or

ruturn true if myStack is empty





```python
def valid_p(s:str)->bool:
    i = 0 # initiate a pointer
    myStack=[] 
    brackets={'(':')','{':'}','[':']'} # a dictionary to store open:closed brackets pair

    while i <=len(s)-1:
        if not myStack: # if mystack is empty, we add s[i] to stack
            myStack.append(s[i])

        else: # if not empty, check if it is a open or close brackets
            if s[i] in brackets: # it is a open bracket
                 myStack.append(s[i])
            else: # for closed, if we check if it matches stack's top
                if s[i]==brackets[myStack[-1]]:
                    myStack.pop()
                else:
                    return False
        # move pointer for append cases
        i+=1

    return len(myStack)==0 # if stack is empty then it is valid
```

```python
print(valid_p('()[]{}'))
print(valid_p('([][]{}'))
```
