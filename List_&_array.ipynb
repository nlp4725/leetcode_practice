# 136 simple number

 Description:
 Given a non-empty array of integers, every element appears twice except for one. Find that single one.
Note:M
Your algorithm should have a linear runtime complexity. Could you implement it without using extra memory?

Example :
Input: [2,2,1] Output: 1



# Thoughts:

# solution 1 hash tabel time and space O(N)

```python
def find_simple_number(nums):
  # skip checking for empty
  # use hash table
  myDict={}
  # iterate the list
  for i in nums: # time O(N) space O(N)
    myDict[i]=myDict.get(i,0)+1
  # solution1: sort it in place dict.sort() (O(NlogN))and return the top
  # find value in dictionary time O(N) space:no (not by key)
  for k,v in myDict.items():
    if v==1:
      return k # guarantee there is one answer
```

```python
find_simple_number([3,3,4])
```

# solution 2 bit operation Time O(N) space O(1)

2^2=0 self ^ self cancels out

0^2=2 self ^ zer0 get kept

2^1^2=1 this can be 2^2^1=0^1=1 means all pairs get canceld out and singel left behind

so single number get left out

```python
def find_simple_number_bit(nums):
  # skip checking for empty
  res=0
  # iterate the list
  for num in nums:
    res=num^res
  return res
```

# Majority element

Given an array of size n, find the majority element. The majority element is the element that appears more than ⌊ n/2 ⌋ times.

Note:
You may assume that the array is non-empty and the majority element always exist in the array.

Example 1:
Input: [3,2,3]

Output: 3



Thoughts:

is empty: no

always answer: yes and one

Hash table: both time and space O(N)

Solution: hash table; time and space: O(N)

```python
def find_majority_element(nums):
  # initiate a dict
  myDict={}
  # iterate the list
  for num in nums: # time O(N) space O(N)
    myDict[num]=myDict.get(num,0)+1
  # find the apperance of the element apperance > n/2 n is  the length
  n=len(nums)
  # iterate the dictionary
  result=0
  for k,v in myDict.items(): # time wise O(N)
    if v>(n//2):
      result=k
  return result
```

```python
find_majority_element([9,8,9])
```

# 1 Two sum

Given an array of integers, return indices of the two numbers such that they add up to a specific target.

Note:
You may assume that each input would have exactly one solution, and you may not use the same element twice.

Example:
Given nums = [2, 7, 11, 15], target = 9,

Because nums[0] + nums[1] = 2 + 7 = 9, return [0, 1]

Thoughts:

is empty: needs check

there is answer: yes one answer?

two solution:
([1,2,3,0],3) ->(0,1) or (2,3)

duplicate
([1,1,4],5) ->(0,2) or (1,2)

target is twice of the element
([2,3,1],6) ->

complexity

hash table time and space O(N)


```python
def two_sum(nums,target):
  """
  type nums: List[int]
  type target: int
  type return: tuple(int,int) or None
  """
  # check if empty
  if not nums:
    return ()
  # initiate a dictionary
  myDict={}
  # iterate the list by index i, save target-current(9-5=4) as key, current's index(0) as value
  for i in range(len(nums)):
    if i == 0: # first one we need to add it in the dictionary
      k=target-nums[i]
      myDict[k]=0
      print('1', myDict)
    else:
        # meanwhile, check if current (4) is in dict's key, then return (dictionary[key], current's index i). time O(N) space O(N)
        # this is important to avoid target = 2* element
      if nums[i] in myDict.keys():
        print('2', myDict)
        return (myDict[nums[i]],i)
      else:
        k=target-nums[i]
        myDict[k]=myDict.get(k,0) + 1
        print('3', myDict)
  return () # make sure this indecention is correct
```

```python
# another algorithm 2*O(N) but still O(N)
# save element and index enumarate() as k,v in dictionary (means no duplicate) O(N)
# iterate the list again, look up the if key = target-element O(N)
```

```python
two_sum([2,3,1],6) # target = element*2, code is correct
```

```python
two_sum([1,2,3,0],3) # two answers, but code generates one
```

```python
two_sum([1,1,4],5) # duplicates, but code generates one
```

# 219-Contains Duplicate II

Description:
Given an array of integers and an integer k,find out whether there are two distinct indices i and j

in the array such that nums[i] = nums[j] and the absolute difference between i and j is at most k.

Example1:
Input: nums = [1,2,3,1], k = 3

Output: true

Example2:
Input: nums = [1,0,1,1], k = 1

Output: true

Example3:
Input: nums = [1,2,3,1,2,3], k = 2

Output: false

Thoughts:

check empty: yes

one answer? maybe just need True or False


sorted: no.

need to sort? not sure O(NlogN)

complexicity:

Brute force: time O(N^2) space O(1)


1. two pointers i j;

2. i loop through range(len(nums))

> for each i, j loop through range(i+1, i+k+1) e.g. i=0 k=2 (1,3) -> 1,2

> > check if nums[i]== nums[j]

> > > return True


## (not correct) Hash Table: time O(N^2) space O(N)

1. loop through nums, (element, index) as (k,v) in a dictionary, v is List[int]

2. for len(v) > 1, have more than 1 indexes:

> > check if |v[o]-v[1]| >= k this assumes only one answer [1,2,1] k=1 when it is [1,1,1,1,1] {1: [0,1,2,3,4]} O(N2) check to see if diff(index) >=k

> > > return True

## correct Hash Table time O(N) space O(N) difference: check as you add to dictionary

1. loop through nums, (element, index) as (k,v) in dicionary

2. if k not in dictionary:

> > add it

> else:

> > check if index - currentv <=k -> return true

> > > else: update the v # we want to keep index difference small







```python
from typing import List

def contains_duplicate(nums: List[int],k:int) -> bool:

  if not nums: # check if empty list or only have 1 element
    return False
  if k<0: # make sure k is positive
    return False
  myDict={}
  for i,e in enumerate(nums): # i: index as value; e: element as key
    if e not in myDict:
      myDict[e]=i
    else:
      if abs(i-myDict[e]) <=k:
        return True
      else:
        myDict[e]=i
  return False # this will handle 1 element or not exist
```

```python
contains_duplicate([],1) # empty, correct
```

```python
contains_duplicate([1],3) # only one element, correct
```

```python
contains_duplicate([1,1],3) # 2 elements
```

```python
contains_duplicate([1,1],1) # 2 elements
```

```python
contains_duplicate([1,2,3,4,2,1],4) # 2 elements
```

Description:
Given two strings s and t , write a function to determine if t is an anagram of s.

Note:
You may assume the string contains only lowercase alphabets.

Example1:

Output: true

Example2:
Input: s = "rat", t = "car"

Thoughts:

complexity:

# using Hash table: time space O(max(m or n)) m, n is the length of s or t s

loop trough s, (char, apperances) as (k,v) in a dictionary1

loop trough t, (char, apperances) as (k,v) in a dictionary2

check if dict1 == dict 2 # can i check directly or sort then check? -> return True

# using 1 Hash Table to save memory

check empty and length

1. loop trough s, (char, apperances) as (k,v) in a dictionary1

2. loop trough t, for char in t,

> > if not in dic1 (something does not match)

> > > return False

> > else

> > > decrease value count in dict1 by 1: dict1[char]-=1

3. check if v in dict1 == 0, if not return False

4. return True






```python
def valid_anagram(s: str,t:str) -> bool:
  if not s or not t or len(s)!=len(t): # check if either is empty or not same length
    return False
  myDict_s={}
  myDict_t={}
  for char in s:
    myDict_s[char]=myDict_s.get(char,0) + 1
  for char in t:
    myDict_t[char]=myDict_t.get(char,0) + 1

  return myDict_s==myDict_t


```
