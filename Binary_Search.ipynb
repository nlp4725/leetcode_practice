# 704 binary search

Given an array of integers nums which is sorted in ascending order, and an integer target, write a function to search target in nums. If target exists, then return its index. Otherwise, return -1.

You must write an algorithm with O(log n) runtime complexity.



Example 1:

Input: nums = [-1,0,3,5,9,12], target = 9
Output: 4
Explanation: 9 exists in nums and its index is 4
Example 2:

Input: nums = [-1,0,3,5,9,12], target = 2
Output: -1
Explanation: 2 does not exist in nums so return -1

Thoughts:

check empty: yes

sorted array? yes

algorithm:

Binary Search O(logN) O(1)

1. get length of array n=len(nums) [0,1,3,4,6,8] target is 6 n =6

2. initiate left_p = 0, right_p=len-1 left=0, right = 5,

while lef <= right: (<= for last element)

3. split in half mid = left_p+ (right-p-left_p)/2 mid = 2

4. if mid == target, return mid

if mid < target, target is on the right half, move left_p to mid+1, righ_p stays the same

if mid > target, target is on the left half,

5. return -1 not found



```python
def binary_search(nums,target):
  if not nums:
    return -1
  left_p=0
  right_p=len(nums)-1
  while left_p<=right_p: # <= vs <: with =, will cover when left_p=4, right_p=4 like last element is the target
    mid = left_p + (right_p-left_p)//2 # mid = (right_p-left_p)//2 problem: right_p = 4, left_p=3  mid =0-> does not work
    # overflow if (right+left)/2
    print('mid',mid)
    if nums[mid]== target:
      return mid
    elif nums[mid] < target: # target is on the right half, move left_p
      left_p =mid +1
      print('left',left_p)
      print('right',right_p)
    else: # target on the left half, move right_p
      right_p =mid-1
      print('left',left_p)
      print('right',right_p)
  return -1



```

# 35. Search Insert Position

Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.

You must write an algorithm with O(log n) runtime complexity.



Example 1:

Input: nums = [1,3,5,6], target = 5
Output: 2
Example 2:

Input: nums = [1,3,5,6], target = 2
Output: 1
Example 3:

Input: nums = [1,3,5,6], target = 7
Output: 4

Thoughts:

empty? check

duplicate? no

sorted? yes

one answer? yes

algorithm


binary serach O(logN) O(1)

1. two points left_p=0, right_p=len-1
2. while left_p<=right_p:

> calculate mid=left_p + ((right_p-left_p)//2)

> check if nums[mid]==target: -> return mid

> else if nums[mid]>target, target on the left half, decrease right_p to mid-1

> else: target on the right half, increase left_p to mid+1

3. if not found, retur left


```python
from typing import List

def search_insertion(nums: List[int],target:int) ->int:
  if not nums: # check if empty array
    return 0 # should return 0
  left_p=0
  right_p=len(nums)-1
  while left_p<=right_p:
    mid=left_p+(right_p-left_p)//2
    if nums[mid]== target:
      return mid
    elif nums[mid] > target: # target on the right
      right_p=mid-1
    else: # target on the left
      left_p=mid+1
  return left_p
```

```python
search_insertion([1,4,6,9],6) # answer: 2
```

```python
search_insertion([1,4,6,9],7) # answer: 3
```

Example 2:

Thoughts:

algorithm

left_p=0, right_p=n-1

while left<=right: (need to double check)

isBadversion(mid)==True -> first appearance is on the left half or mid itself, decrease right_p to mid or mid-1

isBadversion(mid)==False -> first appearance is on the right half, increase left to mid+1

example: [T,T,T,T] -> first apperance is zero



```python
def first_bad_version(versions:List[bool])->int:
  n=len(versions) # simulate a list of array: booleans [F,T,T,T,T] or [T,T,T,T]
  if n==0: # if empty, no result
    return -1

  if versions[-1] ==False: # if last element is not True, all False, no result
    return -1

  left_p=0
  right_p=n-1

  while left_p<right_p: # with = it is infinite loop because right_p=mid, right_p=mid=left_p
    mid=left_p+(right_p-left_p)//2
    if versions[mid]: # if mid is true, first T must be itself or on the left
      right_p=mid
    else:
      left_p=mid+1 # if mid is False, first T must be on the right
  return right_p # why? because last step possible solution is mid/right_p
```

# 349. Intersection of Two Arrays

"Given two integer arrays nums1 and nums2, return an array of their intersection. Each element in the result must be unique and you may return the result in any order.



Example 1:

Input: nums1 = [1,2,2,1], nums2 = [2,2]
Output: [2]
Example 2:

Input: nums1 = [4,9,5], nums2 = [9,4,9,8,4]
Output: [9,4]
Explanation: [4,9] is also accepted."

Thoughts:

answer: potentionally none no intersection [] or List[int]

sorted: no

sorted important?: no

algorithm:

#No need for hash tables with key, vlaues:Two hash tables time and space 0(m+n))

1. num1: (element,index) to (k,v) in dic1
2. num2: (element,index) to (k,v) in dic2
3. for k in dict1:
> > if k in dict2:
> > > add to a set
4. return list(set)

# just use set (hash table with key) O(m+n)

1. num1-> set1

2. num2-> set2

3. intersect(set1, set2)

# two pointers in sorted arrays 0(max(m,n)) not sorted O(mlogm+nlogn)

1) if not sorted, sort first: 0(mlogm+nlogn)

2) initiate two pointers, p1 and p2 at indext 0. initiate a set 

3) while p1 < len(num1) & p2 < len(num2):

 > if num1[p1] == num2[p2], add to a set, move both pointers to the right

 > if num1[p1] < num2[p2], move p1 to the right +1 , p2 stays

 > if num1[p1] > num2[p2], move p2 to the right +1, p1 stays

 return set






```python
# set method
def intersect_two_arrays(num1: List[int],num2: List[int])->List[int]:
  if not num1 or not num2: # check empty
    return []
  
  # convert array to set
  set1=set(num1)
  set2=set(num2)

  # return the intersect
  return list(set1 & set2)

```

```python
# two pointer method
from typing import List

def intersect_two_pointer_method(num1:List[int],num2:List[int]) -> List[int]:
    # sort the array
    num1_sorted=sorted(num1)
    num2_sorted=sorted(num2)

    # initiate two pointers
    p1=0
    p2=0

    # initiate a set to save the results
    result=set()

    # compare arrays and move pointers
    while p1<len(num1) and p2<len(num2):
        if num1_sorted[p1]==num2_sorted[p2]: # if equal, add to the set and move both pointers
            result.add(num1_sorted[p1]) # use add to add element
            p1+=1
            p2+=1
        elif num1_sorted[p1]< num2_sorted[p2]: # if not equal, only move the smaller number's pointer
            p1+=1
        else:
            p2+=1

    return list(result)
```

# Leetcode 367-Valid Perfect Square

"Given a positive integer num, write a function which returns True if num is a perfect square else False. Example1:
Input: num = 16

Output: true

Example2:
Input: num = 14

Output: false
"

Thoughts:

negative? no

0? no

sorted? yes, well it is not array, but can think as one

answer: yes: True or False

algorithm

Binary search (no need to sort: O(nlogn))

1) two pointers left =0 right=num target = num

while left < right or <= (depending on how to move pointers)

2) mid = left+(right-left)//2

if mid**mid == num:

> return True

if mid**mid < num: result is on the right half
> move left pointer to mid+1

if mid**mid > num: result is on the left half
> move right pointer to mid-1

3) if not found, exiwhile loop and return False


```python
from typing import List

def find_sqr(num:int)->bool:
    left=0
    right=num

    while left<=right: # because we move pointers to mid+1 or mid-1, we can use =
        mid=left+(right-left)//2

        result=mid*mid
        if result==num:
            return True
        elif result<num: # need to search right
            left=mid+1
        else:
            right=mid-1
    
    return False

```

#74 search 2d array

"You are given an m x n 2-D integer array matrix and an integer target.

Each row in matrix is sorted in non-decreasing order.
The first integer of every row is greater than the last integer of the previous row.
Return true if target exists within matrix or false otherwise.

Can you write a solution that runs in O(log(m * n)) time?"

Input: matrix = [[1,3,5,7],[10,11,16,20],[23,30,34,60]], target = 3
Output: true

Thoughts:

empty: maybe 

one answer: yes: T or False

sorted? yes. It says non-decreasing -> ascending or flat

algorithm: 

# binary search through 2-D Time O(log(m*n)) space O(1)

Think it as a giant sorted array, indexing it with 2-d indexing.

1) start with 2 pointers l=0, r=m*n-1 (m is # of rows, n is # of cols) 

2) while l<=r:(<= here not < because pointers move to mid+1 or mid-1 not mid):

> > mid=l+(r-l)//2 

>> convert mid to 2-d indexing
   rows=mid//n (n is # of cols) (e.g. 3*3 array, mid of 4 is row 1, col 1)
   cols=mid%n 
>> nums[rows][cols] == target: return True

>> nums[rows][cols] > target: target is on the left half, right pointer need to move to left mid-1

>> numc[rows][cols] < target: targte is on the right half, left poiner need to move to right mid +1

3) return False if not found

# less efficient: Convert to 1-D + binary search. Time(O(m) + Olog(m*n)) Space O(m*n)

1) convert 2-d to 1-d nums. row1,row2,row3,...using + (time O(m) space O(m*n))

2) binary search for the array time O(log(m*n)) space O(1)

left=0
right=len(nums)-1

while left<=right (< or <= depends on how to move pointers. if move to mid+1 mid-1, we can include =, if move to mid, we can not include =)

mid=left+(right-left)//2

if nums[mid] == target:
> return true
elif nums[mid] > target: need to search left half
> move right pointer to mid-1
else nums[mid] < target: need to search right half
> move left pointer to mid+1

exit while loop return False





```python
from typing import List

def searchMatrix(nums:List[List[int]],target:int)->bool:
    if not nums: # check if empty
        return False
    
    m=len(nums)# rows
    n=len(nums[0]) # columns
    
    left=0
    right=m*n-1

    while left<=right: # because we move pointers to mid+1 or mid-1, we can use =
        mid=left+(right-left)//2

        # convert mid to 2-d index
        rows=mid//n
        cols=mid%n
        result=nums[rows][cols]
        # compare to target
        if result==target:
            return True
        elif result<target: # need to search right
            left=mid+1
        else:
            right=mid-1
    
    return False


```
