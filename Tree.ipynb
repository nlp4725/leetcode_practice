# Leetcode 104 - Maximum Depth of Binary Tree

Given a binary tree, find its maximum depth.

The maximum depth is the number of nodes along the longest path from the root node down to the farthest leaf node.


Thought process:

We traverse the tree and record the depth it has seen so far -> local variable

a global variable to record at the final leaf, the max depth among all the leavies

return

def dfs(root,depth_so_far):

    at leaf node
         
        update the global varaible: self.max_depth=max(depth_so_far, max.depth)

    if there is left node child:

        traverse left root, update the depth +1 ; dfs(root.left,depth_so_far+1)

    if there is right node child:

        traverse right root, update the depth +1 ; dfs(root.right,depth_so_far+1)

def max_depth(self,root):

    initiate the max_depth as global variable

    dfs(root,1) depth_so_far=1

    return max_depth


        




```python
# time O(N) space O(1)

class Solution():
    def dfs(self,root,depth_so_far):
        if not root.left and not root.right:
            self.max_depth=max(self.max_depth,depth_so_far)
            return # return at leaf node
        if root.left: # does not return at non-leaf node
            self.dfs(root.left,depth_so_far+1)
        if root.right:
            self.dfs(root.right,depth_so_far+1)

    def max_depth(self,node):
        
        self.max_depth=0

        if not node:
            return self.max_depth
        depth_so_far=1

        self.dfs(node,depth_so_far)

        return self.max_depth
solution=Solution()
solution.max_depth(nodeA)
```

```python
# time O(N) space O(N)

class Solution():
    def dfs(self,root,depth):
        if not root.left and not root.right:
            self.depth_list.append(depth) # when it reaches a leaf, record it to a depth_list
        
        if root.left:
            self.dfs(root.left,depth+1) # current depth +1 is next child's depth
        if root.right:
            self.dfs(root.right,depth+1)

    def maxDepth(self,root):
        if not root:
            return 0
        current_depth=1 # we need a varaible to record current node's status -> depth,but it does not need to be rememberd here
        self.depth_list=[] # a list to track all the leaf nodes' final status
        self.dfs(root,current_depth)

        return max(self.depth_list)

solution=Solution()
solution.maxDepth(nodeA)
```

# Leetcode 257 - Binary Tree Paths

Given a binary tree, return all root-to-leaf paths.


Though process:

Traverse the tree DFS

Record the value at each node when going down in a local variable

At the leaf node,record all the node's values along the path to a global variable 

a gloabl variable to return the final paths 

- initiated as attribute of the class

- updated on leaf node

- use append for updating list


a local variable to track what we have seen so far -> 

- passed an argument for dfs, 

- update as you pass down to each tree

- used + for list 



```python
class Solution(object):
    def dfs(self,root,node_seen_so_far): # 
        if not root.left and not root.right:
            self.node_list.append(node_seen_so_far) 
            return # always return from the base case
        if root.left:
            self.dfs(root.left,node_seen_so_far+'->'+f'{root.left.value}')
        if root.right:
            self.dfs(root.right,node_seen_so_far+'->'+f'{root.right.value}')

        
    def path(self,root):
        if not root:
            return ''
        self.node_list=[] # a global variable is initiated here; a local varaible to track nodes_seen_so_far 
        node_seen_so_far=f'{root.value}'
        self.dfs(root,node_seen_so_far) # 
        return self.node_list

solution=Solution()
solution.path(nodeA)
```

98. Validate Binary Search Tree

Given the root of a binary tree, determine if it is a valid binary search tree (BST).

A valid BST is defined as follows:

The left subtree of a node contains only nodes with keys strictly less than the node's key.
The right subtree of a node contains only nodes with keys strictly greater than the node's key.
Both the left and right subtrees must also be binary search trees.


```python
# Definition for a binary tree node.
#   class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution():
    def dfs(self,root,min_val,max_val):
        if not root.left and not root.right: # leaf node check if it is within the limit
            return root.val>min_val and root.val<max_val
        
        if root.left:
            # check if min< max
            if min_val>max_val:
                return False
            else:
                self.dfs(root.left,min_val,root.val)
                return True # for left node, parent node is bigger than it, then update the max to the parent node
        if root.right:
            # check if min< max
            if min_val>max_val:
                return False
            else:
                self.dfs(root.right,root.val,max_val)
                return True 
            
    def isValidBST(self,root):
        if not root:
            return False
        # carry a min and max at local level
        min_val=-float('inf')
        max_val=float('inf')

        return self.dfs(root,min_val,max_val)
    
```

```python
from collections import deque
# diameter BFS
def get_width(root):
    if not root:
        return
    
    queue=deque([root])
    max_width=0


    while queue: # while there are still node
        level_size=len(queue) # each time the nodes on the the same layer gets add 
        max_width=max(max_width,level_size)
        
        for _ in range(level_size): # for all the nodes on the same layer

            node=queue.popleft() # pop one by one left to right, and add its children node
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
    return max_width

get_width(nodeA)
```
